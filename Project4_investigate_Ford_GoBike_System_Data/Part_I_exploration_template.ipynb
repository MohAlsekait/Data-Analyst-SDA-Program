```python
df['member_age'] = 2019 - df['member_birth_year']
df.head()
```

```python
df['duration_min'] = df['duration_sec'] / 60
df.head()
```

```python
df['duration_hours'] = df['duration_min'] / 60
df.head()
```

```python
df.start_time = pd.to_datetime(df.start_time)
df.end_time = pd.to_datetime(df.end_time)
# source https://towardsdatascience.com/converting-data-to-a-numeric-type-in-pandas-db9415caab0b
#df['member_birth_year'] = df['member_birth_year'] = pd.to_numeric(df['member_birth_year'], errors='coerce').fillna(0).astype('int')
#df['member_age'] =  df['member_age'] = pd.to_numeric(df['member_age'], errors='coerce').fillna(0).astype('int')
```

```python
#extract a lot of data from start time column
df['start_date'] = df.start_time.dt.strftime('%Y-%m-%d')
df['start_hours'] = df.start_time.dt.strftime('%H')
df['start_days'] = df.start_time.dt.strftime('%A')
df['start_month'] = df.start_time.dt.strftime('%B')
```

## Univariate Exploration

```python
def Xcountplot(figsiz,axs,data_df,xdata,rot,tit,xl,yl):
    base_color = sb.color_palette()[0]
    fig = plt.figure(figsize=(figsiz[0], figsiz[1]))
    ax = fig.add_axes(axs)
    sb.countplot(data = data_df, x = xdata, color=base_color);
    plt.xticks(rotation=rot);
    plt.title(tit, y=1.05, fontsize=15)
    plt.xlabel(xl, labelpad=15)
    plt.ylabel(yl, labelpad=15)
    
```

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
Xmonths = pd.api.types.CategoricalDtype(ordered = True, categories = months)
df.start_month = df.start_month.astype(Xmonths)

Xcountplot([13,9],[.125, .125, .775, .755],df,'start_month',60,'Count per start month', 'Start month', 'Count')
```

> as we noticed that the start month of the trip are all at the same month which is feb and that was surprising

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
Xdays = pd.api.types.CategoricalDtype(ordered = True, categories = days)
df.start_days = df.start_days.astype(Xdays)

Xcountplot([13,9],[.125, .125, .775, .755],df,'start_days',60,'Start of week days per count','Days of week', 'Count')
```

> as we noticed here that the weekdays have a higher trip number than the weekend days which proves that a lot of people use it for work or school purposes

```python
df.start_hours = df.start_hours.astype('int')
Xcountplot([15,11],[.125, .125, .775, .755],df,'start_hours',0,'Start hours per count','Start hours','Count')
```

> we can notice that people like the service at the beginning of the morning mostly 8 clocks to the end of the day work which is 15 which proof the theory of what is the purpose of being mostly used by people

## Bivariate Exploration



```python
def countplot(df,x1,h,tit,xl,yl):
    plt.figure(figsize=(15,9))
    sb.countplot(data = df, x = x1, hue = h)
    plt.title(tit, fontsize=22, y=1.015)
    plt.xlabel(xl, labelpad=16)
    plt.ylabel(yl, labelpad=16)
```

```python
 countplot(df, 'start_days', 'member_gender', 'Weekdays per counts for genders', 'Weekdays', 'Count')
```

> we can see here that all genders prefer using the service on Thursday compared to the other days

```python
countplot(df, 'start_days', 'user_type', 'Weekdays per counts for user types', 'Weekdays', 'Count')
```

> We can notice the difference between both of the user types, but we can see that both of the user types increase for Thursday, unlike Subscriber which increases on the weekdays and decreases at the weekend, the Customer  has kind of stable rate for all days

```python
countplot(df, 'member_gender', 'user_type', 'Genders per counts for user types', 'Gender', 'Count')
```

> We can noticed that all genders prefer Subscriber as user type.

```python
countplot(df, 'user_type', 'bike_share_for_all_trip', 'Genders per counts for bike_share_for_all_trip', 'Gender', 'Count')
```

> We noticed that all user type dont prefer sharing bike for all trip.

```python
countplot(df, 'member_gender', 'bike_share_for_all_trip', 'User type per counts for bike_share_for_all_trip', 'User type', 'Count')
```

> We noticed that all genders dont prefer sharing bike for all trip 

```python
subscriber_age = df[df['bike_id']!=0].groupby(['start_hours', 'member_gender']).agg({'bike_id':'count'}).reset_index()

plt.figure(figsize=(16,10))
sb.pointplot(x='start_hours', y='bike_id', hue='member_gender', scale=.6, data=subscriber_age)
plt.title("Start Hours per count for genders", fontsize=22, y=1.015)
plt.xlabel('Start Hours', labelpad=16)
plt.ylabel('Count', labelpad=16)
```

> we can notice here that all genders start from zero or close to it then it reaches the max count two times in a day which is 8 and 15 which is probably the work hours over thier, and between these two times it has a high stable time, but before 8 and after 15 the count goes down

```python
subscriber_age = df[df['bike_id']!=0].groupby(['start_hours', 'user_type']).agg({'bike_id':'count'}).reset_index()

plt.figure(figsize=(16,10))
sb.pointplot(x='start_hours', y='bike_id', hue='user_type', scale=.6, data=subscriber_age)
plt.title("Start Hours per count for User types", fontsize=22, y=1.015)
plt.xlabel('Start hours', labelpad=16)
plt.ylabel('Count', labelpad=16)
```

> we can notice here that all user types start from zero or close to it then it reaches the max count two times in a day which is 8 and 15 which is probably the work hours over thier, and between these two times it has a high stable time, but before 8 and after 15 the count goes down

```python
plt.figure(figsize=(15,10))
base_color = sb.color_palette()[0]
sb.barplot(data=df, x='user_type', y='duration_min', color=base_color);
plt.title("Mean of Duration in Minutes per user type", fontsize=22, y=1.015)
plt.xlabel('User type');
plt.ylabel('Mean of Duration in Minute');
```

> We can see here the mean duration in minutes of customer are higher than the subscriber even though it has the least number, and this tills that customer use the service with more duration than subscriber

```python
plt.figure(figsize=(15,10))
base_color = sb.color_palette()[0]
sb.barplot(data=df, x='member_gender', y='duration_min', color=base_color);
plt.title("Mean of Duration in Minutes per Genders", fontsize=22, y=1.015)
plt.xlabel('Meber genders');
plt.ylabel('Mean of Duration in Minute');
```

> We can see here the mean duration in minutes of others are higher than the females  and males
